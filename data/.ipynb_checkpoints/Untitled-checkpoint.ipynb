```python
import json 
from random import randint

with open('./bike.json', 'r') as file:
    json_data = json.load(file)
    for json_array in json_data:
        for item in json_array:
            item['total'] = randint(0, 7860)
            item['temperatura'] = randint(0, 78)
            item['sensacion'] = randint(0, 100)
        
        #print(item["total"])                      
              
with open('./mybyke.json', 'w') as file:
    json.dump(json_data, file, indent=2)
```

```python
import json 
from random import randint

with open('./bike.json', 'r') as file:
    json_data = json.load(file)

def iterate_multidimensional(my_dict):
    for k,v in my_dict.items():
        if(isinstance(v,dict)):
            print(k+":")
            iterate_multidimensional(v)            
            continue
        print(k+" : "+str(v))
        print("\n")
        
iterate_multidimensional(json_data)    
```
